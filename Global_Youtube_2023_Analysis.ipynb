```python
top_5_videos = df.nlargest(5, 'uploads')[['Youtuber', 'uploads']] #Top 5 Youtube Channels with highest upload 2023
print("Top 5 YouTube channels by number of videos published:")
print(top_5_videos)
```

```python
top_3_subscribers = df.nlargest(3, 'subscribers')[['Youtuber', 'subscribers']] #Top 3 channels by number of subscribers
print("Top 3 YouTube channels by number of subscribers:")
print(top_3_subscribers)
```

```python
richest_youtubers = df.nlargest(10, 'highest_yearly_earnings')[['Youtuber', 'highest_yearly_earnings']] #Richest youtubers in 2023
print("The richest youtubers in 2023 are :")
print(richest_youtubers)
```

```python
df1 = df.groupby('Country')['Youtuber'].count().reset_index()
top_countries = df1.nlargest(10, 'Youtuber')
plt.figure(figsize=(12, 6))
plt.bar(top_countries['Country'], top_countries['Youtuber'])
plt.xlabel('Country')
plt.ylabel('Number of YouTubers')
plt.title('Top 10 Countries with the Highest Number of YouTubers')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
